# Step 1b: CSV to Parquet Conversion
This notebook converts the massive yearly ER visits CSV files into highly compressed Parquet files via chunking.
This enables out-of-core processing using DuckDB for the entire country without saturating RAM.

```python
import pandas as pd
import os
import glob
import pyarrow as pa
import pyarrow.parquet as pq

raw_data_dir = '../data/raw/Atenciones de Urgencia/Atenciones de urgencia de la red pública de Salud/'
processed_data_dir = '../data/processed/urgencias_parquet/'
os.makedirs(processed_data_dir, exist_ok=True)

csv_files = glob.glob(os.path.join(raw_data_dir, '*.csv'))
print(f"Found {len(csv_files)} CSV files to convert.")

for csv_path in csv_files:
    basename = os.path.basename(csv_path).replace('.csv', '.parquet')
    parquet_path = os.path.join(processed_data_dir, basename)
    
    if not os.path.exists(parquet_path):
        print(f"Converting {os.path.basename(csv_path)} to Parquet in chunks...")
        
        # Read in chunks to save memory (500k rows at a time)
        chunksize = 500_000
        parquet_writer = None
        
        try:
            for i, chunk in enumerate(pd.read_csv(csv_path, sep=';', encoding='latin1', chunksize=chunksize, low_memory=False)):
                # Convert IDs to string to prevent mixed type issues
                if 'IdEstablecimiento' in chunk.columns:
                    chunk['IdEstablecimiento'] = chunk['IdEstablecimiento'].astype(str)
                
                table = pa.Table.from_pandas(chunk)
                
                if i == 0:
                    parquet_writer = pq.ParquetWriter(parquet_path, table.schema)
                
                parquet_writer.write_table(table)
                print(f"  Processed chunk {i+1}...")
                
            if parquet_writer:
                parquet_writer.close()
            print(f"Successfully created {basename}\n")
        except Exception as e:
            print(f"Error converting {basename}: {e}\n")
    else:
        print(f"{basename} already exists, skipping.")

print("All conversions completed!")
```
